```python
#imprt the library
import pandas as pd
```

```python
#create dataframe
data={ "Date": [
        "2024-01-05", "2024-01-10", "2024-02-12",
        "2024-02-18", "2024-03-03", "2024-03-15"
    ],
    "Product": ["Laptop", "Mobile", "Laptop", "Tablet", "Mobile", "Laptop"],
    "Price": [50000, 20000, 52000, 15000, 21000, 51000],
    "Quantity": [2, 3, 1, 4, 2, 1]
     }
df=pd.DataFrame(data)
```

```python
#show dataframe
df
```

```python
# showing detailed info about the dataset
df.info()
```

```python
# need to convert Date column from 'object' type to 'DateTime'
df['Date']=pd.to_datetime(df['Date'])
```

```python
df.info()
```

## calculating most selled product

```python
product_sales=df.groupby('Product')['Quantity'].sum()
product_sales
```

```python
# most selled product
product_sales.idxmax()
```

## Total revenue
revenue is quantity*price

```python
#calculating revenue and adding it as a new column
df['Revenue']=df['Price']*df['Quantity']
df
```

```python
# grouping the products and calculating product-wise revenue
df.groupby('Product')['Revenue'].sum()
```

```python
#total revenue
df['Revenue'].sum()
```
