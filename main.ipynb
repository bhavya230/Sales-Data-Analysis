```python
#imprt the library
import pandas as pd
```

```python
#create dataframe
data={ "Date": [
        "2024-01-05", "2024-01-10", "2024-02-12",
        "2024-02-18", "2024-03-03", "2024-03-15"
    ],
    "Product": ["Laptop", "Mobile", "Laptop", "Tablet", "Mobile", "Laptop"],
    "Price": [50000, 20000, 52000, 15000, 21000, 51000],
    "Quantity": [2, 3, 1, 4, 2, 1]
     }
df=pd.DataFrame(data)
```

```python
#show dataframe
df
```

```python
# showing detailed info about the dataset
df.info()
```

```python
# need to convert Date column from 'object' type to 'DateTime'
df['Date']=pd.to_datetime(df['Date'])
```

```python
df.info()
```

## calculating most selled product

```python
product_sales=df.groupby('Product')['Quantity'].sum()
product_sales
```

```python
# most selled product
product_sales.idxmax()
```

## Total revenue
revenue is quantity*price

```python
#calculating revenue and adding it as a new column
df['Revenue']=df['Price']*df['Quantity']
df
```

```python
# grouping the products and calculating product-wise revenue
df.groupby('Product')['Revenue'].sum()
```

```python
#total revenue
df['Revenue'].sum()
```

## Monthly sales analysis

```python
#extract month
df['Month']=df['Date'].dt.month
df
```

```python
# grouping the data month-wise
#summing up monthwise revenue
# and then sorting the monthly revenue in descending order

monthly_revenue=df.groupby('Month')['Revenue'].sum().sort_values(ascending=False)
monthly_revenue
```

```python
## Highest revenue month
monthly_revenue.idxmax()
```

```python
## highest revenue day
df.loc[df['Revenue'].idxmax()]
```

# More Complex Dataset and Problems

```python
data = {
    "Date": [
        "2024-01-05","2024-01-10","2024-01-15",
        "2024-02-12","2024-02-18","2024-02-25",
        "2024-03-03","2024-03-10","2024-03-15",
        "2024-04-02","2024-04-08","2024-04-20",
        "2024-05-05","2024-05-12","2024-05-25"
    ],
    "Product": [
        "Laptop","Mobile","Tablet",
        "Laptop","Tablet","Mobile",
        "Mobile","Laptop","Tablet",
        "Laptop","Mobile","Tablet",
        "Mobile","Laptop","Tablet"
    ],
    "Price": [
        50000,20000,15000,
        52000,15000,21000,
        22000,51000,14000,
        53000,23000,16000,
        24000,55000,17000
    ],
    "Quantity": [
        2,3,4,
        1,2,2,
        3,1,5,
        2,2,3,
        1,2,4
    ],
    "City": [
        "Delhi","Mumbai","Delhi",
        "Bangalore","Delhi","Mumbai",
        "Delhi","Bangalore","Mumbai",
        "Delhi","Mumbai","Bangalore",
        "Delhi","Mumbai","Bangalore"
    ]
}

df = pd.DataFrame(data)

df["Date"] = pd.to_datetime(df["Date"])
df["Revenue"] = df["Price"] * df["Quantity"]

df
```

## Revenue per city

group the dataframe by cities ( all same citiies groups would be created) using groupby() method ,
then sum the on the revenue column using sum()

```python
city_revenue=df.groupby('City')['Revenue'].sum()
```

```python
city_revenue
```

## Best Performing City
City with the Highest Revenue -- sort in descending order of (revenue_per_city)

```python
best_city=city_revenue.sort_values(ascending=False).idxmax()

best_city
```

## least Selling Product

```python
#grouping the products and adding quantity
product_quantity=df.groupby('Product')['Quantity'].sum()

product_quantity
```

```python
#finding least selling product
product_quantity.idxmin()
```

## Which Product Sells Best In Each City

here we are talking about which product is most popular -- the answer of this question we would get by analysing the quantity.

```python
#Group by City and Product --- Adding Qauntity 

product_city_quantity=df.groupby(['City','Product'])['Quantity'].sum()

product_city_quantity
```

```python
# reset index -- gives dataframe
product_city_quantity=product_city_quantity.reset_index()

product_city_quantity
```

```python
# finding each city most selled product quantity 'index no'

# group by city and find the index no where quantity is max -- using idxmax() function

product_city_quantity.groupby('City')['Quantity'].idxmax()
```

```python
# Get the combination where the most selled occured per city

best_per_city=product_city_quantity.loc[product_city_quantity.groupby('City')['Quantity'].idxmax()]

best_per_city
```
